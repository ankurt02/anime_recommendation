```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import precision_score, recall_score, f1_score, mean_absolute_error
from sklearn.cluster import KMeans
import numpy as np

# Fill missing values
df['Tags'] = df['Tags'].fillna('')
df['Studios'] = df['Studios'].fillna('')
df['Type'] = df['Type'].fillna('Unknown')
df['Number Votes'] = pd.to_numeric(df['Number Votes'], errors='coerce').fillna(0)
df['Rating Score'] = df['Rating Score'].fillna(df['Rating Score'].mean())

# Normalize numerical features
scaler = StandardScaler()
scaled_votes = scaler.fit_transform(df[['Number Votes']])
scaled_ratings = scaler.fit_transform(df[['Rating Score']])

# Vectorize 'Tags'
tag_vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
tfidf_matrix = tag_vectorizer.fit_transform(df['Tags']).toarray()

# Encode 'Studios' and 'Type'
studio_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
studios_encoded = studio_encoder.fit_transform(df[['Studios']])

type_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
type_encoded = type_encoder.fit_transform(df[['Type']])

features_combined = np.hstack([
    scaled_votes,
    scaled_ratings,
    tfidf_matrix,
    studios_encoded,
    type_encoded
])

num_clusters = 12

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    features_combined, 
    df['Rating Score'], 
    test_size=0.2, 
    random_state=42
)

model = KMeans(n_clusters=num_clusters, random_state=42)
model.fit(X_train)

y_pred = model.predict(X_test)

median_rating = df['Rating Score'].median()
y_test_binary = (y_test > median_rating).astype(int)
y_pred_binary = (y_pred > np.median(y_pred)).astype(int)

precision = precision_score(y_test_binary, y_pred_binary)
recall = recall_score(y_test_binary, y_pred_binary)
f1 = f1_score(y_test_binary, y_pred_binary)

mae = mean_absolute_error(y_test, y_pred)

print("Model Evaluation Metrics:")
print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1 Score: {f1:.3f}")
print(f"Mean Absolute Error: {mae:.3f}")

df['Cluster'] = model.predict(features_combined)

def get_recommendations(title, n_recommendations=5):
    try:
        idx = df[df['Name'].str.lower() == title.lower()].index[0]
        
        cluster = df.loc[idx, 'Cluster']
        similar_anime = df[df['Cluster'] == cluster]
        recommendations = similar_anime[similar_anime.index != idx].nlargest(n_recommendations, 'Rating Score')
        
        return recommendations[['Name', 'Rating Score', 'Tags']]
    
    except IndexError:
        return f"Anime '{title}' not found in database."
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Example usage
anime_title = "Naruto"
print(f"\nRecommendations for {anime_title}:")
print(get_recommendations(anime_title))
```

### Eculidean Distance

```python
from scipy.spatial.distance import cdist
from sklearn.metrics import precision_score, recall_score, f1_score, mean_absolute_error
import numpy as np

def recommend_anime_euclidean(title, top_n=10):
    anime_index = get_index_from_title(title)
    anime_features = features_combined[anime_index]
    
    # Euclidean distances
    distances = cdist([anime_features], features_combined, metric='euclidean').flatten()
    
    
    closest_indices = np.argsort(distances)[1:top_n + 1]
    recommended_anime = [get_title_from_index(idx) for idx in closest_indices]
    
    
    y_pred = distances[closest_indices]
    y_true = df.iloc[closest_indices]['Rating Score'].values
    
    
    median_rating = np.median(df['Rating Score'])
    y_true_binary = (y_true > median_rating).astype(int)
    y_pred_binary = (y_pred < np.median(y_pred)).astype(int) 
    
    precision = precision_score(y_true_binary, y_pred_binary)
    recall = recall_score(y_true_binary, y_pred_binary)
    f1 = f1_score(y_true_binary, y_pred_binary)
    mae = mean_absolute_error(y_true, y_pred)
    
    print("\nModel Evaluation Metrics:")
    print(f"Precision: {precision:.3f}")
    print(f"Recall: {recall:.3f}") 
    print(f"F1 Score: {f1:.3f}")
    print(f"Mean Absolute Error: {mae:.3f}")
    
    return recommended_anime


anime_liked = "Naruto"
recommended_anime_euclidean = recommend_anime_euclidean(anime_liked)
print(f"\nAnime similar to '{anime_liked}' based on Euclidean distance:")
for anime in recommended_anime_euclidean:
    print(anime)
```

#### Manhattan Distance

```python
def recommend_anime_manhattan(title, top_n=10):
    anime_index = get_index_from_title(title)
    anime_features = features_combined[anime_index]
    
    # Calculate Manhattan distances
    distances = cdist([anime_features], features_combined, metric='cityblock').flatten()
    
    closest_indices = np.argsort(distances)[1:top_n + 1]
    recommended_anime = [get_title_from_index(idx) for idx in closest_indices]
    
    y_pred = distances[closest_indices]
    y_true = df.iloc[closest_indices]['Rating Score'].values
    
    median_rating = np.median(df['Rating Score'])
    y_true_binary = (y_true > median_rating).astype(int)
    y_pred_binary = (y_pred < np.median(y_pred)).astype(int) 
    
    precision = precision_score(y_true_binary, y_pred_binary)
    recall = recall_score(y_true_binary, y_pred_binary)
    f1 = f1_score(y_true_binary, y_pred_binary)
    mae = mean_absolute_error(y_true, y_pred)
    
    print("\nModel Evaluation Metrics:")
    print(f"Precision: {precision:.3f}")
    print(f"Recall: {recall:.3f}") 
    print(f"F1 Score: {f1:.3f}")
    print(f"Mean Absolute Error: {mae:.3f}")
    
    return recommended_anime

recommended_anime_manhattan = recommend_anime_manhattan(anime_liked)
print(f"\nAnime similar to '{anime_liked}' based on Manhattan distance:")
for anime in recommended_anime_manhattan:
    print(anime)
```

#### Minkowski Distance

```python
def recommend_anime_minkowski(title, p=3, top_n=10):
    anime_index = get_index_from_title(title)
    anime_features = features_combined[anime_index]
    
    # Calculate Minkowski distances
    distances = cdist([anime_features], features_combined, metric='minkowski', p=p).flatten()
    
    closest_indices = np.argsort(distances)[1:top_n + 1]
    recommended_anime = [get_title_from_index(idx) for idx in closest_indices]
    
    return recommended_anime

recommended_anime_minkowski = recommend_anime_minkowski(anime_liked, p=3)
print(f"Anime similar to '{anime_liked}' based on Minkowski distance:")
for anime in recommended_anime_minkowski:
    print(anime)
```
